```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

```

```python
warnings.filterwarnings("ignore")
```

```python
pd.set_option("display.max_columns", None)
```

```python
final_df = pd.read_csv("cleaned.csv")
```

```python
final_df.drop("Unnamed: 0", axis =1, inplace=True)
```

```python
final_df["GDP/Capita"] = final_df["GDP"]/final_df["Population Total"]
```

```python
final_df["Energy Usage/Capita"] = final_df["Energy Usage"] / final_df["Population Total"]
final_df["CO2 Emissions/Capita"] = final_df["CO2 Emissions"] / final_df["Population Total"]
final_df["Life Expectancy Combined"] = (final_df["Life Expectancy Female"] + final_df["Life Expectancy Male"])/2
```

```python
all_countries_mean = final_df.groupby(['Country']).mean()
```

```python
all_countries_mean["Tourism Status"] = all_countries_mean["Tourism Inbound"] - all_countries_mean['Tourism Outbound']
```

```python
all_countries_mean["Tourism Status"] = np.where(all_countries_mean["Tourism Status"] > 0, "Inbound", np.where(all_countries_mean["Tourism Status"] < 0, "Outbound", "Balanced"))
```

```python
countries = final_df.Country.unique().tolist()
```

```python
all_countries_mean
```

```python
df = all_countries_mean[["Birth Rate",'Business Tax Rate','Days to Start Business'	,'Health Exp % GDP',
                         'Health Exp/Capita','Hours to do Tax','Infant Mortality Rate','Internet Usage','Lending Interest',
                         'Mobile Phone Usage',	'Population 0-14','Population 15-64','Population 65+','Population Urban',
                         'GDP/Capita','Energy Usage/Capita',	'CO2 Emissions/Capita',	'Life Expectancy Combined',
                         'Tourism Status']]
```

```python
from sklearn.preprocessing import RobustScaler, StandardScaler
```

```python
sc = StandardScaler()
rs = RobustScaler()
```

```python
df = pd.get_dummies(df)
```

```python
scaled_sc = sc.fit_transform(df)
scaled_rs = rs.fit_transform(df)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering , DBSCAN
```

```python
from sklearn.metrics import silhouette_score 
```

```python
wcss_sc = []
sil_score_sc = []
wcss_rs = []
sil_score_rs = []
for i in range(2,20):
    km = KMeans(n_clusters=i)
    labels = km.fit_predict(scaled_sc)
    wcss_sc.append(km.inertia_)
    sil_score_sc.append(silhouette_score(scaled_sc, labels))
    
    labels = km.fit_predict(scaled_rs)
    wcss_rs.append(km.inertia_)
    sil_score_rs.append(silhouette_score(scaled_rs, labels))
    
    
plt.figure(figsize=(10,8))
plt.subplot(221 ) 
plt.plot(range(2,20), wcss_sc, marker = "o", linestyle="-", color="white", )
plt.title("wcss standard scaler")

plt.subplot(222) 
plt.plot(range(2,20), wcss_rs, marker = "o", linestyle="-", color="white")
plt.title("wcss robust scaler")

plt.subplot(223)
plt.plot(range(2,20), sil_score_sc, marker = "o", linestyle="-", color="red")
plt.title("silhouette score standard scaler")

plt.subplot(224)
plt.plot(range(2,20), sil_score_rs, marker = "o", linestyle="-", color="red")
plt.title("silhouette score robust scaler")


plt.tight_layout()
```

```python
from sklearn.decomposition import PCA
```

```python
pca_sc = PCA()
pca_rs = PCA()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(121)
vecotrs_sc = pca_sc.fit_transform(scaled_sc)
plt.plot(pca_sc.explained_variance_ratio_.cumsum(), marker="o")
pca_sc.explained_variance_ratio_.cumsum()

plt.subplot(122)
vecotrs_rs = pca_rs.fit_transform(scaled_rs)
plt.plot(pca_rs.explained_variance_ratio_.cumsum(), marker="o")
pca_rs.explained_variance_ratio_.cumsum();
```

```python
wcss_pca_sc = []
sil_score_pca_sc = []
wcss_pca_rs = []
sil_score_pca_rs = []

for k in range(2, 20):
    km = KMeans(n_clusters=k)
    labels_sc = km.fit_predict(vecotrs_sc[:,:11])
    sil_score_pca_sc.append(silhouette_score(vecotrs_rs[:,:11], labels_sc))
    wcss_pca_sc.append(km.inertia_)
    
    km = KMeans(n_clusters=k)
    labels_rs = km.fit_predict(vecotrs_rs[:,:2])
    sil_score_pca_rs.append(silhouette_score(vecotrs_rs[:,:2], labels_rs))
    wcss_pca_rs.append(km.inertia_)
    
    
    
plt.figure(figsize=(10,8))
plt.subplot(221) 
plt.plot(range(2,20), wcss_pca_sc, marker = "o", linestyle="-", color="blue")
plt.title("wcss pca standard scaler")

plt.subplot(222) 
plt.plot(range(2,20), wcss_pca_rs, marker = "o", linestyle="-", color="blue")
plt.title("wcss pca robust scaler")

plt.subplot(223)
plt.plot(range(2,20), sil_score_pca_sc, marker = "o", linestyle="-", color="red")
plt.title("silhouette pca score standard scaler")

plt.subplot(224)
plt.plot(range(2,20), sil_score_pca_rs, marker = "o", linestyle="-", color="red")
plt.title("silhouette pca score robust scaler")


plt.tight_layout()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
categorical_columns = ['Tourism Status']

numerical_transformer = RobustScaler()
categorical_transformer = OneHotEncoder(drop='first')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, all_countries_mean.columns.difference(['year'] + categorical_columns)),
        ('cat', categorical_transformer, categorical_columns)
    ])

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('kmeans', KMeans(n_clusters=3))
])

all_countries_mean = all_countries_mean.drop(columns=['year'])

pipeline.fit(all_countries_mean)

cluster_labels = pipeline.predict(all_countries_mean)
```
