# COLUMNS DESCRIPTION

- **Birth Rate**	The number of live births per thousand of population per year.
- **Business Tax Rate (%)**	A levy placed by government to a firm's profit
- **CO2 Emissions**	Carbon dioxide emission per capita
- **Country**	Country name
- **Days to Start Business**	The number of calendar days needed to complete the procedures to legally operate a business
- **Energy Usage**	Energy consumption of a country
- **GDP ($)**	Gross Domestic Product
- **Health Exp \% GDP**	Percentage of GDP spent on health care
- **Health Exp/Capita**	Health expenditure per capita
- **Hours to do Tax**	The time, in hours per year, it takes to prepare, file, and pay (or withhold) three major types of taxes
- **Infant Mortality Rate**	the number of deaths per 1,000 live births of children under one year of age
- **Internet Usage**	Percentage of population that using internet
- **Lending Interest**	The proportion of a loan that is charged as interest to the borrower, typically expressed as an annual percentage of the loan outstanding.
- **Life Expectancy Female**	The average period that a female may expect to live.
- **Life Expectancy Male**	The average period that a male may expect to live.
- **Mobile Phone Usage**	Percentage of mobile phone users
- **Population 0-14**	Population of 0-14 years of age
- **Population 15-64**	Population of 15-64 years of age
- **Population 65+**	Population of 65 years of age
- **Population Total**	Total of overall population
- **Population Urban**	Total of urban population
- **Tourism Inbound**	The tourism spending of visitors who are not residents of that country
- **Tourism Outbound**	The tourism spending of residents of certain country to another
- **Year**	Year

```python
wcss_sc = []
sil_score_sc = []
wcss_rs = []
sil_score_rs = []
for i in range(2,20):
    km = KMeans(n_clusters=i)
    labels = km.fit_predict(scaled_sc)
    wcss_sc.append(km.inertia_)
    sil_score_sc.append(silhouette_score(scaled_sc, labels))
    
    labels = km.fit_predict(scaled_rs)
    wcss_rs.append(km.inertia_)
    sil_score_rs.append(silhouette_score(scaled_rs, labels))
    
    
plt.figure(figsize=(10,8))
plt.subplot(221 ) 
plt.plot(range(2,20), wcss_sc, marker = "o", linestyle="-", color="blue")
plt.title("wcss standard scaler")

plt.subplot(222) 
plt.plot(range(2,20), wcss_rs, marker = "o", linestyle="-", color="blue")
plt.title("wcss robust scaler")

plt.subplot(223)
plt.plot(range(2,20), sil_score_sc, marker = "o", linestyle="-", color="red")
plt.title("silhouette score standard scaler")

plt.subplot(224)
plt.plot(range(2,20), sil_score_rs, marker = "o", linestyle="-", color="red")
plt.title("silhouette score robust scaler")


plt.tight_layout()
```

```python
def get_graph_through_tsne(data, scalar="standard", perplexity_values=[10, 15, 20,25, 30, 40], 
                           n_iters=[1000,2000,3000, 5000], model=None, graph=False):

    if scalar == "standard":
        scalar_obj = StandardScaler()
        scaled_data = scalar_obj.fit_transform(data)

    if scalar == "robust":
        scalar_obj = RobustScaler()
        scaled_data = scalar_obj.fit_transform(data)

    silhouette = []
    cluster = []
    model_params_outer = []
    perpexility_outer = []
    n_iter_outer = []
    scalar_name_outer = []  
    dimensionality_reduction_method = [] 
    algorithm_name_outer = []

    
    count = 1
    for iters in n_iters:
        
        for perpexility in perplexity_values:
            silhouette_inner = []
            wcss_inner = []
            model_params_inner = []

            for k in range(2, 10):

                tsne = TSNE(n_components=2, random_state=42, n_iter=iters, perplexity=perpexility)
                tsne_result = tsne.fit_transform(scaled_data)

                if model is None or model == 'kmeans':
                    model = KMeans(n_clusters=k)
                    
                elif model == 'agglomerative':
                    model = AgglomerativeClustering(n_clusters=k)
                    
                else:
                    model_params = model.get_params()
                    model_params["n_clusters"] = k
                    model = model.set_params(**model_params)
                    

                model_params_inner.append(model.get_params())
                labels = model.fit_predict(tsne_result)
                silhouette_inner.append(silhouette_score(tsne_result, labels))

            cluster_max = sorted(enumerate(silhouette_inner), key=lambda x: x[1], reverse=True)[0][0]+2
            cluster.append(cluster_max)
            for i in model_params_inner:
                if i["n_clusters"] == cluster_max:
                    model_params_outer.append(i)
                else: pass
            max_silhouette = sorted(enumerate(silhouette_inner), key=lambda x: x[1], reverse=True)[0][1]
            silhouette.append(max_silhouette)
            perpexility_outer.append(perpexility)
            n_iter_outer.append(iters)
            scalar_name_outer.append(scalar)  
            dimensionality_reduction_method.append('tsne')
            algorithm_name_outer.append(model.__class__.__name__) 
            
            if graph:
                subplot_count = len(n_iters) * len(perplexity_values)

                plt.figure(figsize=(5, subplot_count * 5))
                plt.subplot(subplot_count, 1, count)
                plt.plot(range(2, 10), silhouette_inner, marker="o", linestyle="-", color="red")
                plt.axhline(max(silhouette_inner), label=max(silhouette_inner), linestyle="--", color="grey")
                plt.title(f"{iters} iters, {perpexility} perplexity, {scalar} scalar")
                plt.legend()
                count += 1

    if graph:
        plt.tight_layout()
        plt.show()

    result_df = pd.DataFrame({
        'Silhouette': silhouette,
        'Cluster': cluster,
        'Model_Params': model_params_outer,
        'perplexity': perpexility_outer,
        'n_iter': n_iter_outer,
        'Scalar': scalar_name_outer,
        'Algorithm': algorithm_name_outer,
        'Dimensionality_Reduction': dimensionality_reduction_method  
    })
    
    
    return result_df
```

```python

def get_graph_through_pca(data, scalar="standard", model=None, graph=False):

    if scalar == "standard":
        scalar_obj = StandardScaler()
        scaled_data = scalar_obj.fit_transform(data)

    if scalar == "robust":
        scalar_obj = RobustScaler()
        scaled_data = scalar_obj.fit_transform(data)

    silhouette = []
    cluster = []
    model_params_outer = []
    n_components_outer = []
    scalar_name_outer = []  
    dimensionality_reduction_method = [] 
    algorithm_name_outer = []

    count = 1

    pca = PCA()
    pca_result = pca.fit_transform(scaled_data)
    variance = pca.explained_variance_ratio_.cumsum()
    n_components_to_keep = next(i for i, value in enumerate(variance) if value >= 0.94)

    for k in range(2, 10):
        if model is None or model == 'kmeans':
            model = KMeans(n_clusters=k)

        elif model == 'agglomerative':
            model = AgglomerativeClustering(n_clusters=k)

        else:
            model_params = model.get_params()
            model_params["n_clusters"] = k
            model = model.set_params(**model_params)

        labels = model.fit_predict(pca_result[:, :n_components_to_keep])
        silhouette_score_value = silhouette_score(pca_result[:, :n_components_to_keep], labels)

        cluster_max = k
        best_model_params = model.get_params()

        if silhouette_score_value > max(silhouette, default=-1):
            cluster_max = k
            best_model_params = model.get_params()

        cluster.append(cluster_max)
        n_components_outer.append(n_components_to_keep)
        model_params_outer.append(best_model_params)
        silhouette.append(silhouette_score_value)
        scalar_name_outer.append(scalar)  
        dimensionality_reduction_method.append('pca')
        algorithm_name_outer.append(model.__class__.__name__) 

        if graph:
            plt.figure(figsize=(5, 5))
            plt.plot(range(2, 10), silhouette, marker="o", linestyle="-", color="red")
            plt.axhline(max(silhouette), label=max(silhouette), linestyle="--", color="grey")
            plt.title(f"{n_components_to_keep} components, {k} clusters, {scalar} scalar")
            plt.legend()

    if graph:
        plt.tight_layout()
        plt.show()

    result_df = pd.DataFrame({
        'Silhouette': silhouette,
        'Cluster': cluster,
        'Model_Params': model_params_outer,
        'n_components': n_components_outer,
        'Scalar': scalar_name_outer,
        'Algorithm': algorithm_name_outer,
        'Dimensionality_Reduction': dimensionality_reduction_method  
    })
    return result_df
```

```python
results = pd.DataFrame(columns=["Silhouette", "Cluster", "Model_Params", "perplexity", "n_iter",
                     "Scalar", "Algorithm", "Dimensionality_Reduction"])

for scalar in ["robust", "standard"]:
    
    for linkage in ["complete", "ward", "average", "single"]: ####### AGGLOMERATIVE
        model = AgglomerativeClustering(linkage=linkage)
        
        df = get_graph_through_tsne(all_countries_mean.drop("year",axis=1),scalar,model=model)
        results = pd.concat([results,df])
        
        df = get_graph_through_pca(all_countries_mean.drop("year",axis=1),scalar,model=model)
        results = pd.concat([results,df])
        
        
    for algorithm in ["elkan", "lloyd"]:
        model = KMeans(algorithm=algorithm)
        
        df = get_graph_through_tsne(all_countries_mean.drop("year",axis=1),scalar,model=model)
        results = pd.concat([results,df])
        
        df = get_graph_through_pca(all_countries_mean.drop("year",axis=1),scalar,model=model)
        results = pd.concat([results,df])
    
```

```python

correlation_threshold = 0.75
corr_matrix = all_countries_mean.drop("year", axis=1).corr().abs()
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
highly_correlated = upper_triangle[upper_triangle > correlation_threshold].stack()

features_to_remove = set()

for idx in highly_correlated.index:
    features_to_remove.add(idx[0] if corr_matrix.loc[idx] > corr_matrix.loc[(idx[1], idx[1])] else idx[1])

subset_features = [col for col in all_countries_mean.columns if col not in features_to_remove]

subset_df = all_countries_mean[subset_features]
subset_df
```
