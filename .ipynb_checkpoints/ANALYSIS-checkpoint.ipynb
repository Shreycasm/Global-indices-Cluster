```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
final_df = pd.read_csv("cleaned.csv")
```

## TOP  AND BOTTOM 10 GDP AND GLOBAL GDP ACROSS YEARS

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
top_10_gdp = (final_df.groupby("Country")["GDP"].mean()).sort_values(ascending=False).head(10)
ax[0].bar(top_10_gdp.index, top_10_gdp)
ax[0].spines["top"].set_color("w")
ax[0].set_xticklabels(top_10_gdp.index,rotation =90)

for i , value in enumerate(top_10_gdp.head(10)):
    ax[0].text(i,value ,f'{value:.1e}', ha='center', va='bottom', rotation =90)

    
    
bottom_10_gdp = (final_df.groupby("Country")["GDP"].mean()).sort_values(ascending=True).head(10)
ax[1].bar(bottom_10_gdp.index, bottom_10_gdp)
ax[1].spines["top"].set_color("w")
ax[1].set_xticklabels(bottom_10_gdp.index,rotation =90)

for i , value in enumerate(bottom_10_gdp.head(10)):
    ax[1].text(i,value ,f'{value:.1e}', ha='center', va='bottom', rotation =90)    

    
    
plt.figure(figsize =(16,4))
plt.plot(final_df.groupby('year')["GDP"].sum().index,final_df.groupby('year')["GDP"].sum(),  marker="o", linestyle="-") ;
```

## TOP AND BOTTOM 10 POPULATION AND GLOBAL POPULATION ACROSS YEARS

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
top_10_gdp = (final_df.groupby("Country")["Population Total"].mean()/1e6).sort_values(ascending=False).head(10)
ax[0].bar(top_10_gdp.index, top_10_gdp)
ax[0].spines["top"].set_color("w")
ax[0].set_xticklabels(top_10_gdp.index,rotation =90)

for i , value in enumerate(top_10_gdp.head(10)):
    ax[0].text(i,value ,f'{value:.2f}', ha='center', va='bottom', rotation =90)

    
    
bottom_10_gdp = (final_df.groupby("Country")["Population Total"].mean()/1e6).sort_values(ascending=True).head(10)
ax[1].bar(bottom_10_gdp.index, bottom_10_gdp)
ax[1].spines["top"].set_color("w")
ax[1].set_xticklabels(bottom_10_gdp.index,rotation =90)

for i , value in enumerate(bottom_10_gdp.head(10)):
    ax[1].text(i,value ,f'{value:.2f}', ha='center', va='bottom', rotation =90)    

    
    
plt.figure(figsize =(16,4))
plt.plot(final_df.groupby('year')["Population Total"].sum().index,final_df.groupby('year')["Population Total"].sum(),  marker="o", linestyle="-") ;
```

## Checking population an gdp of top and bottom countries 

```python
filtered_countries = final_df[final_df["Country"].isin(["United States", 'China', 'San Marino', "Kiribati"])]
mean = filtered_countries.groupby("Country")[["Population Total", "GDP"]].mean()
mean.plot(kind ="bar", log=True);
```

## GDP PER CAPITA

```python
final_df['GDP/Capita'] = final_df["GDP"] / final_df["Population Total"]
```

```python
filtered_countries = final_df[final_df["Country"].isin(["United States", 'China', 'San Marino', "Kiribati"])]
mean = (filtered_countries.groupby("Country")[["Population Total", "GDP", "GDP/Capita"]].mean())
ax = mean.plot(kind ="bar", log=True ,figsize=(6, 4))
ax.spines["top"].set_color("w")

ax.legend(bbox_to_anchor =(1,1), loc="upper left")
for i in ax.containers:
    ax.bar_label(i,rotation=90,fmt='%.1e', label_type='edge')
```

## TOP AND BOTTOM 10 GDP/CAPITA COUNTRY AND GDP/CAPITA OVER YEARS GOBAL

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
top_10_gdp = (final_df.groupby("Country")["GDP/Capita"].mean()).sort_values(ascending=False).head(10)
ax[0].bar(top_10_gdp.index, top_10_gdp)
ax[0].spines["top"].set_color("w")
ax[0].set_xticklabels(top_10_gdp.index,rotation =90)

for i , value in enumerate(top_10_gdp.head(10)):
    ax[0].text(i,value ,f'{value:.2f}', ha='center', va='bottom', rotation =90)

    
    
bottom_10_gdp = (final_df.groupby("Country")["GDP/Capita"].mean()).sort_values(ascending=True).head(10)
ax[1].bar(bottom_10_gdp.index, bottom_10_gdp)
ax[1].spines["top"].set_color("w")
ax[1].set_xticklabels(bottom_10_gdp.index,rotation =90)

for i , value in enumerate(bottom_10_gdp.head(10)):
    ax[1].text(i,value ,f'{value:.2f}', ha='center', va='bottom', rotation =90)    

    
    
plt.figure(figsize =(16,4))
plt.plot(final_df.groupby('year')["GDP/Capita"].sum().index,final_df.groupby('year')["GDP/Capita"].sum(),  marker="o", linestyle="-") ;
```

## GDP GROWTH RATE

```python
mean_by_year = final_df.groupby('year')[["GDP", "GDP/Capita", "Population Total"]].mean()
```

```python
mean_by_year["gdp_growth_rate"] = mean_by_year["GDP"].pct_change()*100
mean_by_year["GDP/Capita_grwoth_rate"] = mean_by_year["GDP/Capita"].pct_change()*100
mean_by_year["Population_grwoth_rate"] = mean_by_year["Population Total"].pct_change()*100
```

```python
mean_by_year[['gdp_growth_rate'	,'GDP/Capita_grwoth_rate',	'Population_grwoth_rate']].plot(figsize=(16,4), style="-o")
plt.legend(bbox_to_anchor =(1,1), loc="upper left")
```

## POPULATION OF WORLD BY AGE GROUP

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.pie([final_df["Population 0-14"].mean() , final_df["Population 15-64"].mean(), final_df["Population 65+"].mean()]
      ,colors=["teal", "r", "pink"], wedgeprops={'width':0.2}, autopct="%.2f%%", labels= ["0-14",'15-64','65+'])

plt.subplot(1,2,2)
plt.pie([final_df["Population 0-14"].mean() + final_df["Population 65+"].mean(), final_df["Population 15-64"].mean()]
      ,colors=["w", "r"], wedgeprops={'width':0.2}, autopct="%.2f%%", labels= ["Dependent",'Working Class'])
plt.pie([final_df["Population 0-14"].mean() + final_df["Population 65+"].mean(), final_df["Population 15-64"].mean()]
      ,colors=["teal", "r"], wedgeprops={'width':0.1},radius=0.95,);
```

## Urban vs Rural Population

## life expentacy male vs female

```python
final_df["Life Expectancy Combined"] = (final_df["Life Expectancy Female"] + final_df["Life Expectancy Male"])/2
fig, ax = plt.subplots(3,2, figsize=(15,15))
top_10_gdp = (final_df.groupby("Country")["Life Expectancy Female"].mean()).sort_values(ascending=False).head(10)
ax[0,0].bar(top_10_gdp.index, top_10_gdp,)
ax[0,0].spines["top"].set_color("w")
ax[0,0].set_xticklabels(top_10_gdp.index,rotation =90)
ax[0,0].set_title("Countries Where Womens Lives Longer")
for i , value in enumerate(top_10_gdp.head(10)):
    ax[0,0].text(i,value ,f'{value:.0f}', ha='center', va='bottom',)
plt.tight_layout()
    
    
bottom_10_gdp = (final_df.groupby("Country")["Life Expectancy Female"].mean()).sort_values(ascending=True).head(10)
ax[0,1].bar(bottom_10_gdp.index, bottom_10_gdp)
ax[0,1].spines["top"].set_color("w")
ax[0,1].set_xticklabels(bottom_10_gdp.index, rotation =90)
ax[0,1].set_title("Countries Where Womens Lives Shorter")
for i , value in enumerate(bottom_10_gdp.head(10)):
    ax[0,1].text(i,value ,f'{value:.0f}', ha='center', va='bottom',)
plt.tight_layout()

top_10_gdp = (final_df.groupby("Country")["Life Expectancy Male"].mean()).sort_values(ascending=False).head(10)
ax[1,0].bar(top_10_gdp.index, top_10_gdp,)
ax[1,0].spines["top"].set_color("w")
ax[1,0].set_xticklabels(top_10_gdp.index,rotation =90)
ax[1,0].set_title("Countries Where Male Lives Longer")
for i , value in enumerate(top_10_gdp.head(10)):
    ax[1,0].text(i,value ,f'{value:.0f}', ha='center', va='bottom',)
plt.tight_layout()
    

bottom_10_gdp = (final_df.groupby("Country")["Life Expectancy Male"].mean()).sort_values(ascending=True).head(10)
ax[1,1].bar(bottom_10_gdp.index, bottom_10_gdp)
ax[1,1].spines["top"].set_color("w")
ax[1,1].set_xticklabels(bottom_10_gdp.index, rotation =90)
ax[1,1].set_title("Countries Where Male Lives Shorter")

for i , value in enumerate(bottom_10_gdp.head(10)):
    ax[1,1].text(i,value ,f'{value:.0f}', ha='center', va='bottom')
plt.figure(figsize =(16,4))
plt.tight_layout()


top_10_gdp = (final_df.groupby("Country")["Life Expectancy Combined"].mean()).sort_values(ascending=False).head(10)
ax[2,0].bar(top_10_gdp.index, top_10_gdp,)
ax[2,0].spines["top"].set_color("w")
ax[2,0].set_xticklabels(top_10_gdp.index,rotation =90)
ax[2,0].set_title("Countries with High Life expectancy")
for i , value in enumerate(top_10_gdp.head(10)):
    ax[2,0].text(i,value ,f'{value:.0f}', ha='center', va='bottom',)
plt.tight_layout()
    
    
bottom_10_gdp = (final_df.groupby("Country")["Life Expectancy Combined"].mean()).sort_values(ascending=True).head(10)
ax[2,1].bar(bottom_10_gdp.index, bottom_10_gdp)
ax[2,1].spines["top"].set_color("w")
ax[2,1].set_xticklabels(bottom_10_gdp.index, rotation =90)
ax[2,1].set_title("Countries with Low Life expectancy")
for i , value in enumerate(bottom_10_gdp.head(10)):
    ax[2,1].text(i,value ,f'{value:.0f}', ha='center', va='bottom',)
plt.tight_layout()



final_df.groupby('year')["Life Expectancy Female"].mean().plot(style="--o", label="Life Expectancy Female")
final_df.groupby('year')["Life Expectancy Male"].mean().plot(style="--o", label="Life Expectancy Male")
final_df.groupby('year')["Life Expectancy Combined"].mean().plot(style="--o", label="Life Expectancy Combined")
plt.legend();
```
